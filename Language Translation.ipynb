```python
from transformers import MBartForConditionalGeneration,MBart50TokenizerFast
model=MBartForConditionalGeneration.from_pretrained("facebook/mbart-large-50-many-to-many-mmt")
tokenizer=MBart50TokenizerFast.from_pretrained("facebook/mbart-large-50-many-to-many-mmt",src_lang="en_XX")
```

```python
from tkinter import *
from PIL import ImageTk,Image
to_lang='hi_IN'
root=Tk()
root.title('Language Translation')
root.geometry('800x600')
root.configure(bg='skyblue')
title=Label(root,text='Language Translation',font=("Helvetica",30),fg="black",bg='skyblue')
title.place(x=220,y=5)
input_text=Label(root,text="Enter a english sentence:",font=("Helvetica",15),fg="black",bg='skyblue')
input_text.place(x=5,y=80)
input_entry=Entry(root,borderwidth=5,font=("Helvetica",15))
input_entry.grid(row=5,column=3,ipadx=150,padx=240,pady=85)
select_text=Label(root,text='Select the language:',font=("Helvetica",15),fg="black",bg='skyblue')
select_text.place(x=5,y=150)
lang=StringVar()#declaring variable for 
lang.set('hi')
b1=Radiobutton(root,text='Hindi',variable=lang,value='hi',font=("Helvetica",15),fg="black",bg='skyblue')
b1.place(x=300,y=150)
b2=Radiobutton(root,text='Telugu',variable=lang,value='te',font=("Helvetica",15),fg="black",bg='skyblue')
b2.place(x=300,y=200)
b3=Radiobutton(root,text='Tamil',variable=lang,value='ta',font=("Helvetica",15),fg="black",bg='skyblue')
b3.place(x=300,y=250)
b4=Radiobutton(root,text='German',variable=lang,value='ge',font=("Helvetica",15),fg="black",bg='skyblue')
b4.place(x=500,y=150)
b5=Radiobutton(root,text='French',variable=lang,value='fr',font=("Helvetica",15),fg="black",bg='skyblue')
b5.place(x=500,y=200)
b6=Radiobutton(root,text='Japanese',variable=lang,value='ja',font=("Helvetica",15),fg="black",bg='skyblue')
b6.place(x=500,y=250)
b7=Radiobutton(root,text='Urdu',variable=lang,value='ur',font=("Helvetica",15),fg="black",bg='skyblue')
b7.place(x=300,y=300)
b8=Radiobutton(root,text='Chinese',variable=lang,value='ch',font=("Helvetica",15),fg="black",bg='skyblue')
b8.place(x=500,y=300)
b8=Radiobutton(root,text='Korean',variable=lang,value='ko',font=("Helvetica",15),fg="black",bg='skyblue')
b8.place(x=300,y=350)
b8=Radiobutton(root,text='Russian',variable=lang,value='ru',font=("Helvetica",15),fg="black",bg='skyblue')
b8.place(x=500,y=350)

def language_selected():
    text=lang.get()
    global to_lang
    if text=='te':
        to_lang='te_IN'
    elif text=='hi':
        to_lang='hi_IN'
    elif text=='ge':
        to_lang='de_DE'
    elif text=='ja':
        to_lang='ja_XX'
    elif text=='ta':
        to_lang='ta_IN'
    elif text=='fr':
        to_lang='fr_XX'
    elif text=='ur':
        to_lang='ur_PK'
    elif text=='ch':
        to_lang='zh_CN'
    elif text=='ko':
        to_lang='ko_KR'
    elif text=='ru':
        to_lang='ru_RU'
    sent=input_entry.get()
    model_input=tokenizer(sent,return_tensors='pt')
    generated_tokens=model.generate(**model_input,
                                forced_bos_token_id=tokenizer.lang_code_to_id[to_lang])
    translation=tokenizer.batch_decode(generated_tokens,skip_special_tokens=True)
    trans_entry.delete(0,END)
    trans_entry.insert(0,translation)
    
submit_button=Button(root,text='SUBMIT',font=("Helvetica",15),fg="black",bg='lightgreen',command=language_selected)
submit_button.place(x=300,y=410)
def clear_text():
    input_entry.delete(0,END)
    trans_entry.delete(0,END)
clear_button=Button(root,text='CLEAR',font=("Helvetica",15),fg="black",bg='red',command=clear_text)
clear_button.place(x=470,y=410)
trans_text=Label(root,text='Translatted text:',font=("Helvetica",15),fg="black",bg='skyblue')
trans_text.place(x=5,y=500)
trans_entry=Entry(root,borderwidth=5,font=("Helvetica",15))
trans_entry.grid(row=6,column=3,ipadx=170,pady=295)

root.mainloop()
```
